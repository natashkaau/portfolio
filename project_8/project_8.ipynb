**Подключение к базе и выгрузка данных**

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
path_to_db = '/datasets/telecomm_csi.db' # путь к файлу
```

```python
engine = create_engine(f'sqlite:///{path_to_db}', echo = False)
```

```python
query = """SELECT user.user_id,
                  user.lt_day,
                  CASE
                      WHEN user.lt_day <= 365 THEN TRUE
                      WHEN user.lt_day > 365 THEN FALSE
                  END AS is_new,
                  user.age,
                  CASE
                      WHEN user.gender_segment=1 THEN 'женщины'
                      WHEN user.gender_segment=0 THEN 'мужчины'
                  END AS gender_segment,
                  user.os_name,
                  user.cpe_type_name,
                  location.country,
                  location.city,
                  age_segment.title AS age_segment,
                  traffic_segment.title AS traffic_segment,
                  lifetime_segment.title AS lifetime_segment,
                  user.nps_score,
                  CASE
                      WHEN user.nps_score>=0 AND user.nps_score<=6 THEN 'критики'
                      WHEN user.nps_score>=7 AND user.nps_score<=8 THEN 'нейтралы'
                      WHEN user.nps_score>=9 AND user.nps_score<=10 THEN 'cторонники'
                  END AS nps_group
           FROM user
           LEFT JOIN location ON user.location_id=location.location_id
           LEFT JOIN age_segment ON user.age_gr_id=age_segment.age_gr_id
           LEFT JOIN traffic_segment ON user.tr_gr_id=traffic_segment.tr_gr_id
           LEFT JOIN lifetime_segment ON user.lt_gr_id=lifetime_segment.lt_gr_id
        """
```

```python
df = pd.read_sql(query, engine) 
df
```

```python
df.to_csv('df.csv')
```
